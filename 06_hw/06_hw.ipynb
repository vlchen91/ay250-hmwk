Create an election and prediction table.

```python
connection = sqlite3.connect("election.db")
cursor = connection.cursor()

sql_cmd = "DROP TABLE IF EXISTS election"
cursor.execute(sql_cmd)
sql_cmd = "DROP TABLE IF EXISTS prediction"
cursor.execute(sql_cmd)

sql_cmd1 = """

            CREATE TABLE election (
                dem_cid         INTEGER  NOT NULL  PRIMARY KEY,
                rep_cid         INTEGER,
                dem_volume      INTEGER,
                rep_volume      INTEGER,
                state_name      TEXT,
                electoral_votes INTEGER );
                
           """

sql_cmd2 = """

            CREATE TABLE prediction (
                pid INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT DEFAULT 0,
                dem_cid         INTEGER,
                rep_cid         INTEGER,
                closingdate     DATE,
                dem_price       REAL,
                rep_price       REAL );

           """
cursor.execute(sql_cmd1)
cursor.execute(sql_cmd2)

connection.commit()
connection.close()
```

Populate the prediction table with data.

```python
connection = sqlite3.connect("election.db")
cursor = connection.cursor()

sql_cmd = "SELECT dem_cid,rep_cid FROM election"
cursor.execute(sql_cmd)
db_info = cursor.fetchall()

for entry in db_info:
    response = urllib2.urlopen("http://api.intrade.com/jsp/XML/MarketData/ClosingPrice.jsp?conID=%d"%\
        entry[0])
    html1 = response.read()
    response.close()
    response = urllib2.urlopen("http://api.intrade.com/jsp/XML/MarketData/ClosingPrice.jsp?conID=%d"%\
        entry[1])
    html2 = response.read()
    response.close()
    soup_con1 = BeautifulSoup(html1)
    soup_con2 = BeautifulSoup(html2)
    closingprice1 = soup_con1.find('closingprice')
    closingprice2 = soup_con2.find('closingprice')
    cps1 = closingprice1.find_all('cp')
    cps2 = closingprice2.find_all('cp')
    
    dem_cid = entry[0]
    rep_cid = entry[1]
    
    cps = None
    if not cps1:
        cps = cps2
    if not cps2:
        cps = cps1
    # if both dem (1) and rep (2) data exist...
    if cps1 and cps2:
        if len(cps1) < len(cps2):    
            cps = cps1 # by default, choose cps1
        else:
            cps = cps2
    
    for i in range(len(cps)):
        closingdate = cps[-(i+1)].get('date')
        closingdate = closingdate.split(' ')[1] # Get the %m/%d/%Y date only
        closingdate = closingdate.split('/')
        y = closingdate[2]
        m = closingdate[0]
        d = closingdate[1]
        y_m_d = "20{}-{}-{}".format(y,m,d)
        dem_price = None
        rep_price = None
        if not cps1:
            dem_price = -1.0 # -1.0 is default for NO DATA AVAILABLE
        else:
            dem_price = float(cps1[-(i+1)].get('price'))
        if not cps2:
            rep_price = -1.0
        else:
            rep_price = float(cps2[-(i+1)].get('price'))
        sql_cmd = """ INSERT INTO prediction VALUES (
                      {},{},{},'{}',{},{} ) """.format(
                      "NULL",dem_cid,rep_cid,y_m_d,dem_price,rep_price)
        cursor.execute(sql_cmd)
            
connection.commit()
connection.close()
```

### Part C

Simulate 10,000 elections using the prediction probabilities today. Plot the histogram of electoral college votes for Obama. Compare the fraction of elections won by Obama to today’s general election probability in intrade and at fivethirtyeight.

Repeat this for predictions on July 1, Sept 1, and Oct 1. You should store your simulation results in another db table.

```python
def run_simulation(which='state',n_times=10000,day="NOW','-1 day",verbose=False):
    ''' Run the 2012 Presidential Election over "which" in-trade data
        from "day", "n_times".

        e.g. run_simulation(which="general",n_times=4000,day="2012-04-25")
        will run the simulation using the general election in-trade data
        from 04/25/2012, a total of 4,000 times.

        run_simulation returns (date,obama,romney) for storage into a 
        result SQL table.

        The verbose parameter controls whether or not to display the
        print messages.

        The which parameter can be only "state" or "general".
    '''
    sql_cmd1_lst = [  
                    "SELECT dem_price,rep_price,election.electoral_votes,",
                    "election.dem_volume,election.rep_volume FROM prediction",
                    "JOIN election on election.dem_cid = prediction.dem_cid",
                    "WHERE DATE(closingdate) = DATE('%s')" % day,
                    "AND NOT election.state_name = 'General'"
                   ]
    sql_cmd2_lst = [
                    "SELECT dem_price,rep_price,election.electoral_votes,",
                    "election.dem_volume,election.rep_volume FROM prediction",
                    "JOIN election on election.dem_cid = prediction.dem_cid",
                    "WHERE DATE(closingdate) = DATE('%s')" % day,
                    "AND election.state_name = 'General'"
                   ]
    sql_cmd = None
    if which.lower() == 'general':
        sql_cmd = " ".join(sql_cmd2_lst)
    else:
        sql_cmd = " ".join(sql_cmd1_lst)
        
    connection = sqlite3.connect("election.db")
    cursor = connection.cursor()
    cursor.execute(sql_cmd)
    db_info = cursor.fetchall()
    
    obama_votes = 0
    romney_votes = 0
    for entry in db_info:
        dem_price = entry[0]
        rep_price = entry[1]
        # It doesn't matter if price == -1.0, because the corresponding volume
        # is 0. The threshold variable takes care of this.
        votes = entry[2]
        dem_vol = entry[3]
        rep_vol = entry[4]
        threshold = ( (dem_price * 0.01)*dem_vol + (1 - rep_price * 0.01)*rep_vol ) /\
            (dem_vol + rep_vol)
        for i in range(n_times):
            u = scipy.stats.uniform.rvs()
            if u <= threshold:
                obama_votes += votes
            else:
                romney_votes += votes
    
    obama = int(round(obama_votes/float(n_times)))
    romney = int(round(romney_votes/float(n_times)))
    
    if verbose:
        winner = "Obama" if obama > romney else "Romney"
        print "Using the in-trade data for %s election(s):" % which
        print "-------------------------------------------"
        print "over %d rounds of simulation, on average..." % n_times
        print "Obama scores",obama,"electoral votes."
        print "Romney scores",romney,"electoral votes."
        print "%d + %d = %d" % (obama,romney,obama+romney)
        print winner,"is the winner!"
    
    if day == "NOW','-1 day":
        sql_cmd = "SELECT DATE('%s')" % day
        cursor.execute(sql_cmd)
        day = cursor.fetchall()
        day = day[0][0]
        connection.close()
    
    return str((day,obama,romney))
```

### Part D

The state markets differ from the general election market because they both represent two different types of markets. Ideally, the in-trade general election market price depends on the price of each state, state's voting history, state polling data, etc etc. In short, the general election should be the sum of the 50 states' elections. I believe this is how Nate Silver comes up with his statistics.

However, I posit it is highly unlikely for the buyer/seller of the general election market to scientifically and systematically keep track of every of the 50 states to formulate a confidence percentage (e.g. 60% Obama, 40% Romney). There may be people like that, (e.g. me who's running this simulation), but the in-trade state markets clearly do not indicate that. Overall, the trade volume of the 50 states combined is around 2 orders of magnitude smaller than that of the general election.** The state markets thus do not combine to represent accurately the general election.

**One reason for this: buyers/sellers don't bother trading when they see that the market prices are > 60% or < 40%. Realistically, a state with > 60% chance will win 100% of the time. So they don't trade. There are also outliers that the small volume of the states market doesn't screen out. For example, a pro-Romney trader might put "Romney wins 100%" for all 50 states, and the state markets will reflect that to a significant degree.

Try to alter the simulations in c) such that any state where p<40% or p>60% is fixed for the current favorite.Try weighting by trade volume.

```python
def run_converged_simulation(n_times=10000,day="NOW','-1 day",verbose=False):
    ''' Run the 2012 Presidential Election over state in-trade data
        from "day", "n_times".

        run_simulation returns "(date,obama,romney)" for storage into a 
        result SQL table.

        The verbose parameter controls whether or not to display the
        print messages.

        Converges p>= 60% --> 100%, p<=40% --> 0%
    '''
    sql_cmd_lst = [  
                    "SELECT dem_price,rep_price,election.electoral_votes,",
                    "election.dem_volume,election.rep_volume FROM prediction",
                    "JOIN election on election.dem_cid = prediction.dem_cid",
                    "WHERE DATE(closingdate) = DATE('%s')" % day,
                    "AND NOT election.state_name = 'General'"
                   ]
    sql_cmd = " ".join(sql_cmd_lst)
        
    connection = sqlite3.connect("election.db")
    cursor = connection.cursor()
    cursor.execute(sql_cmd)
    db_info = cursor.fetchall()
    
    obama_votes = 0
    romney_votes = 0
    for entry in db_info:
        dem_price = entry[0]
        rep_price = entry[1]
        # It doesn't matter if price == -1.0, because the corresponding volume
        # is 0. The threshold variable takes care of this.
        votes = entry[2]
        dem_vol = entry[3]
        rep_vol = entry[4]
        threshold = ( (dem_price * 0.01)*dem_vol + (1 - rep_price * 0.01)*rep_vol ) /\
            (dem_vol + rep_vol)
        if threshold >= 0.60:
            obama_votes += votes * 10000
        elif threshold <= 0.40:
            romney_votes += votes * 10000
        else:
            for i in range(n_times):
                u = scipy.stats.uniform.rvs()
                if u <= threshold:
                    obama_votes += votes
                else:
                    romney_votes += votes
        
    obama = int(round(obama_votes/float(n_times)))
    romney = int(round(romney_votes/float(n_times)))
    
    if verbose:
        winner = "Obama" if obama > romney else "Romney"
        print "over %d rounds of simulation, on average..." % n_times
        print "Obama scores",obama,"electoral votes."
        print "Romney scores",romney,"electoral votes."
        print "%d + %d = %d" % (obama,romney,obama+romney)
        print winner,"is the winner!"
    
    if day == "NOW','-1 day":
        sql_cmd = "SELECT DATE('%s')" % day
        cursor.execute(sql_cmd)
        day = cursor.fetchall()
        day = day[0][0]
        connection.close()
        
    return str((day,obama,romney))
```
