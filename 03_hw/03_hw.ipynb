```python
######## PART A ########

def rejection_sampler_1D( num_samples, target_dist, ref_dist,
    target_df=[], ref_df=[] ):
    
    """
    A 1-D sampler using tools from scipy.stats. The sampler takes in as input the
    a scipy.stats function (e.g. scipy.stats.norm), and the number of samples desired.
    It outputs the samples, the value of M, and the proportion of samples that
    was accepted.

    'target_dist'    - the desired scipy.stats module to sample from.
    'ref_dist'       - the reference scipy.stats module.

    'target_df'      - the shape parameter(s) for the target (Default is [])
    'ref_df'         - the shape parameter(s) for the reference (Default is [])

    Example call:
    
        [samples, M, proportion_accepted] = rejection_sampler_1D( 1000, 
            scipy.stats.laplace(0.,1.), scipy.stats.norm(0.,1.) )
    
    """
    # Shape parameter parsing
    if type(target_df) == list:
        f_df = target_df
    else:
        f_df = [] + [target_df]
    if type(ref_df) == list:
        g_df = ref_df
    else:
        g_df = [] + [ref_df]
    
    f = lambda x,args: target_dist.pdf(x,*args)
    g = lambda x,args: ref_dist.pdf(x,*args)
    
    # find M | M g(x) <= f(x) for all x
    ## Design a method such that if g < f for *most* x, increment M by some amount
    
    ## arbitrary default values
    num_above = 1000
    flag = True
    M = 1
    
    while flag:
        f_discrete = f(np.linspace(0.,4.,1000),f_df)
        Mg_discrete = g(np.linspace(0.,4.,1000),g_df) * M
        # ^^^ pdfs return arrays, not lists
        bool_above = f_discrete > Mg_discrete
        if len(bool_above[bool_above == True]) > 0:
            M += 0.05
        else:
            flag = False  
    
    # Sample
    samples = []
    i, accept_count, total_count = 0, 0., 0.
    x_c_rvs_func = lambda params: ref_dist.rvs(*params)
    while i < num_samples:
        # Step 1: Sample a candidate, x_c, from g(x) and u from U(0,1)
        x_c = x_c_rvs_func(g_df)
        u = scipy.stats.uniform.rvs()
        
        # Step 2: if u <= f(x_c) / Mg(x_c), then accept x_c, otherwise reject x_c
        f_sub = f(x_c,f_df)
        Mg_sub = M * g(x_c,g_df)
        if u < f_sub / Mg_sub:
            samples.append(x_c)
            i += 1
            accept_count += 1
        total_count += 1    
    
    ansatz = [samples, M, accept_count / total_count]
    return ansatz
```

```python
######## PART B ########

my_dist = scipy.stats.laplace()
ref_dist = scipy.stats.cauchy()
answer = rejection_sampler_1D(1000,my_dist,ref_dist)
print
print "First 10 samples:\n\t%.2f, %.2f, %.2f, %.2f, %.2f, %.2f, %.2f, %.2f, %.2f, %.2f" % \
    tuple([i for i in sort(answer[0][:10])])
print 'M value:', answer[1]
print 'Percent accepted: %0.2f%%' % (answer[2] * 100)
d_val = scipy.stats.kstest(answer[0],my_dist.cdf)[0]
print 'Kolmogorov-Smirnov D-value:', d_val
print

data = answer[0]
hist(data,100,normed=True,range=(-5,5))
ax = plt.gca()
ax.plot(linspace(-5,5,5000),my_dist.pdf(linspace(-5,5,5000)),linewidth=3.0,
    label='target')
ax.plot(linspace(-5,5,5000),answer[1]*ref_dist.pdf(linspace(-5,5,5000)),
    linewidth=2.0,c='r',label='reference',linestyle='--')
plt.legend()
```
