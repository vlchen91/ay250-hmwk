```python
from scipy import *
import aifc
# Modify the following parameters if you want larger figures :)
mpl.rcParams['figure.figsize'] = 12,9
mpl.rcParams['figure.dpi'] = 80
```

```python
i = 1
# Open an audio file
af = aifc.open('homework4_data/%d.aif'%i,'rb')
RATE = af.getframerate()
data = af.readframes(af.getnframes())
af.close()

integer_data = fromstring(data,dtype=np.int32)
time = arange(size(integer_data)) / float(RATE)

# Perform FFT
freq = np.fft.fftfreq(time.shape[-1], d=RATE**-1)
ft = np.fft.fft(integer_data)
nbins = time.shape[-1]
plt.plot(freq[:nbins/32],abs(ft[:nbins/32]) ** 2)
title('Power Spectral Density')
xlabel('Freq [Hz]')
ylabel('Power')
ind_max = np.where(abs(ft[:nbins/32])**2 == max(abs(ft[:nbins/32])**2))[0][0]
freq_max = freq[ind_max]
print "Max power at",freq_max,"Hz"
    
```
